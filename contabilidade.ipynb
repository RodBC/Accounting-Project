```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#PRIMEIRO FAÇA O UPLOAD DA PLANILHA SIGA Brasil_2018-2022_Valores repassados_Estados_Saúde (Mod. Ap.)
# EM FORMATO CSV
# renomei-e a mesma como "dados.csv"
dt = pd.read_csv("dadoscsv.csv")
# Exibe algumas linhas do DataFrame
dt
```

```python
primeiro_ano = dt.loc[dt['Ano'] == 2018]
qtd_primeiro_ano = primeiro_ano.count()[0]
segundo_ano = dt.loc[dt['Ano'] == 2019]
qtd_segundo_ano = segundo_ano.count()[0]
terceiro_ano = dt.loc[dt['Ano'] == 2020]
qtd_terceiro_ano = terceiro_ano.count()[0]
quarto_ano = dt.loc[dt['Ano'] == 2021]
qtd_quarto_ano = quarto_ano.count()[0]
quinto_ano = dt.loc[dt['Ano'] == 2022]
qtd_quinto_ano = quinto_ano.count()[0]

labels = [f'2018 \n {qtd_primeiro_ano}', f'2019 \n {qtd_segundo_ano}', f'2020 \n{qtd_terceiro_ano}', f'2021 \n{qtd_quarto_ano}', f'2022 \n{qtd_quinto_ano}']
values = [qtd_primeiro_ano, qtd_segundo_ano, qtd_terceiro_ano, qtd_quarto_ano, qtd_quinto_ano ]

fig, ax = plt.subplots(figsize=(10, 7.5))
sns.pointplot(x=labels, y=values)
ax.set_title('Registros por ano')
plt.show()

```

```python
subfuncao_mapping = {
    '302': '302 - ASSISTÊNCIA HOSPITALAR E AMBULATORIAL',
    '304': '304 - VIGILÂNCIA SANITÁRIA',
    '305': '305 - VIGILÂNCIA EPIDEMIOLÓGICA',
    '303': '303 - SUPORTE PROFILÁTICO E TERAPÊUTICO',
    '122': '122 - ADMINISTRAÇÃO GERAL',
    '301': '301 - ATENÇÃO BÁSICA',
    '511': '511 - SANEAMENTO BÁSICO RURAL',
    '512': '512 - SANEAMENTO BÁSICO URBANO',
    '846': '846 - OUTROS ENCARGOS ESPECIAIS',
    '572': '572 - DESENVOLVIMENTO TECNOLÓGICO E ENGENHARIA',
    '422': '422 - DIREITOS INDIVIDUAIS, COLETIVOS E DIFUSOS',
    '571': '571 - DESENVOLVIMENTO CIENTÍFICO',
    '128': '128 - FORMAÇÃO DE RECURSOS HUMANOS'
}

#pegando subfuncao desejada
for i in subfuncao_mapping.items():
  print(i)
desired_subfuncao = input("Qual das subfunções acima você deseja ver? informe o numero apenas: ")
while desired_subfuncao not in subfuncao_mapping:
    print("Numero de subfuncao invalido:")
    desired_subfuncao = input("Qual das subfunções acima você deseja ver: ")

# pegando qntd de  de registros para subfuncao em cada ano

def count_subfuncao_per_year(dt, subfuncao):
    subfuncao_count = dt[dt['Subfunção'] == subfuncao].groupby('Ano').size().reset_index(name='Count')
    count_dict = dict(zip(subfuncao_count['Ano'], subfuncao_count['Count']))
    return count_dict

desired_subfuncao = subfuncao_mapping[desired_subfuncao]
result = count_subfuncao_per_year(dt, desired_subfuncao)

print(f"\nTotal de registros para a subfunção {desired_subfuncao} em cada ano:\n{result}\n")


#pegando tipo de valor aplicado
columns = ['Empenhado', 'Liquidado', 'Pago', 'RP Pago', 'Despesa Executada']
desired_column = input(f"Qual dos valores abaixo \n {columns} \n voce deseja ver para a subfunção escolhida {(desired_subfuncao)}: ")

while desired_column not in columns:
    print("Valor Invalido")
    desired_column = input("Qual dos valores voce deseja ver para a subfunção escolhida {(desired_subfuncao)}:")

#pegando quantidade valorada para determinada subfuncao em cada ano
anos = [2018, 2019, 2020, 2021, 2022, 2018, 2019]
def calculate_total_spent(dt, column, desired_subfuncao):
    results = {}
    for ano in anos: #ja estamos vendo ano a ano
        total_sum = 0
        for value, subfuncao in zip(dt.loc[dt['Ano'] == ano, desired_column].values, dt.loc[dt['Ano'] == ano, 'Subfunção'].values):
            if isinstance(value, str) and subfuncao == desired_subfuncao:
                try:
                    num = pd.to_numeric(value.replace('.', '').replace(',', '.'), errors='raise')
                    total_sum += num
                except (ValueError, TypeError):
                    continue
        formatted_sum = '{:,.2f}'.format(total_sum).replace(',', '.')
        split_value = formatted_sum.split('.')
        formatted_value = '.'.join(split_value[:-1]) + ',' + split_value[-1]
        results[ano] = formatted_value

    return results
valor_total_por_subfuncao_e_tipogasto = calculate_total_spent(dt, desired_column, desired_subfuncao)
print(f"\n Valor total de {desired_column} para a subfunção {desired_subfuncao} em cada ano:")
print(valor_total_por_subfuncao_e_tipogasto)
```
