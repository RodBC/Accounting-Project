```python
# removendo valores inuteis
dt = dt.drop(columns=['Instituição','Cod.IBGE', 'População', 'Identificador da Conta'])
dt
```

```python
primeiro_ano = dt.loc[dt['Ano'] == 2018]
qtd_primeiro_ano = primeiro_ano.count()[0]
segundo_ano = dt.loc[dt['Ano'] == 2019]
qtd_segundo_ano = segundo_ano.count()[0]
terceiro_ano = dt.loc[dt['Ano'] == 2020]
qtd_terceiro_ano = terceiro_ano.count()[0]
quarto_ano = dt.loc[dt['Ano'] == 2021]
qtd_quarto_ano = quarto_ano.count()[0]

labels = ['2018', '2019', '2020', '2021']
values = [qtd_primeiro_ano, qtd_segundo_ano, qtd_terceiro_ano, qtd_quarto_ano ]

fig, ax = plt.subplots(figsize=(10, 7.5))
sns.pointplot(data=dt, x=labels, y=values)
ax.set_title('Quantidade de despesas por ano')
plt.show()
```
