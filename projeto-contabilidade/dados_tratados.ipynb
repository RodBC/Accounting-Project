```python
# Use a função loc para filtrar as linhas com base em vários critérios
filtro = dt.loc[(dt['UF'] == 'SP')]
soma = filtro['Valor (R$)'].sum()
print(f'A soma dos valores na coluna "Valor (R$)" é: {soma}')

filtro
```

```python
def filtro(ano=None, uf=None, coluna=None, conta=None):
    resultado = dt
    if ano:
        resultado = resultado.loc[resultado['Ano'] == ano]
    if uf:
        resultado = resultado.loc[resultado['UF'] == uf]
    if coluna:
        resultado = resultado.loc[resultado['Coluna'] == coluna]
    if conta:
        resultado = resultado.loc[resultado['Conta'].str.startswith(conta)]
        
    return resultado

def recebe_parametros():
    ano = input("Ano (deixe em branco para todos): ")
    if ano:
        ano = int(ano)
    uf = input("UF (deixe em branco para todos): ")
    coluna = input("Coluna (deixe em branco para todos): ")
    conta = input("Conta (deixe em branco para todos): ")
    
    resultado = filtro(ano, uf, coluna, conta)
    return resultado
    


# Exemplo de chamada
resultado = recebe_parametros()
resultado
```

```python
# Somar os valores por 'UF'
resultado_somado = resultado.groupby('UF')['Valor (R$)'].sum().reset_index()

# Plotagem do gráfico de barras
plt.figure(figsize=(10, 6))  # Define o tamanho da figura

# Colunas do DataFrame para os eixos x e y
estados = resultado_somado['UF']
valores_gastos = resultado_somado['Valor (R$)']

# Cria o gráfico de barras
plt.bar(estados, valores_gastos)

# Títulos e rótulos
plt.title('Soma dos Valores Gastos por Estado')
plt.xlabel('Estado')
plt.ylabel('Soma dos Valores Gastos (R$) TUDO EM BILHAO')

# Exibe o gráfico
plt.show()
```
